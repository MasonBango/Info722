```
data = spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/shared_uploads/lucas.guo@wine-searcher.com/Life_Expectancy_Data.csv")
```

```
df = data
```

```
df = df.select("Life expectancy ","Alcohol"," BMI ","Population","Schooling","Total expenditure","GDP","Income composition of resources")
display(df)
```

```
df = df.withColumnRenamed("Life expectancy ","life_expectancy")
df = df.withColumnRenamed(" BMI ","BMI")
df = df.withColumnRenamed("Total expenditure","expenditure")
df = df.withColumnRenamed("Income composition of resources","income")
display(df)
```

```
df = df.withColumn('life_expectancy',df.life_expectancy.cast('double'))
df = df.withColumn('Alcohol',df.Alcohol.cast('double'))
df = df.withColumn('BMI',df.BMI.cast('double'))
df = df.withColumn('Population',df.Population.cast('double'))
df = df.withColumn('Schooling',df.Schooling.cast('double'))
df = df.withColumn('expenditure',df.expenditure.cast('double'))
df = df.withColumn('GDP',df.GDP.cast('double'))
df = df.withColumn('income',df.income.cast('double'))
df.printSchema()
```

```
from pyspark.sql.functions import col,isnan, when, count
display(df.select([count(when(isnan(c) | col(c).isNull(), c)).alias(c) for c in df.columns]))
```

```
df = df.na.drop()
display(df.describe())
```

```
from pyspark.ml.feature import QuantileDiscretizer
qds = QuantileDiscretizer(numBuckets=2,inputCol="life_expectancy", outputCol="life_expectancy_2buckets", relativeError=0.01, handleInvalid="error")
bucketizer = qds.fit(df)
df = qds.setHandleInvalid("keep").fit(df).transform(df)
qds = QuantileDiscretizer(numBuckets=10,inputCol="Alcohol", outputCol="Alcohol_10buckets", relativeError=0.01, handleInvalid="error")
bucketizer = qds.fit(df)
df = qds.setHandleInvalid("keep").fit(df).transform(df)
qds = QuantileDiscretizer(numBuckets=10,inputCol="BMI", outputCol="BMI_10buckets", relativeError=0.01, handleInvalid="error")
bucketizer = qds.fit(df)
df = qds.setHandleInvalid("keep").fit(df).transform(df)
qds = QuantileDiscretizer(numBuckets=10,inputCol="Population", outputCol="Population_10buckets", relativeError=0.01, handleInvalid="error")
bucketizer = qds.fit(df)
df = qds.setHandleInvalid("keep").fit(df).transform(df)
qds = QuantileDiscretizer(numBuckets=10,inputCol="Schooling", outputCol="Schooling_10buckets", relativeError=0.01, handleInvalid="error")
bucketizer = qds.fit(df)
df = qds.setHandleInvalid("keep").fit(df).transform(df)
qds = QuantileDiscretizer(numBuckets=10,inputCol="expenditure", outputCol="expenditure_10buckets", relativeError=0.01, handleInvalid="error")
bucketizer = qds.fit(df)
df = qds.setHandleInvalid("keep").fit(df).transform(df)
qds = QuantileDiscretizer(numBuckets=10,inputCol="GDP", outputCol="GDP_10buckets", relativeError=0.01, handleInvalid="error")
bucketizer = qds.fit(df)
df = qds.setHandleInvalid("keep").fit(df).transform(df)
qds = QuantileDiscretizer(numBuckets=10,inputCol="income", outputCol="income_10buckets", relativeError=0.01, handleInvalid="error")
bucketizer = qds.fit(df)
df = qds.setHandleInvalid("keep").fit(df).transform(df)
```

```
display(df)
```

```
df = df.drop('life_expectancy','Alcohol','BMI','Population','Schooling','expenditure','GDP','income')
display(df.groupBy('Alcohol_10buckets').count())
```

```
from pyspark.ml.feature import (VectorAssembler,VectorIndexer,OneHotEncoder,StringIndexer)
from pyspark.ml import Pipeline
from pyspark.ml.classification import LogisticRegression,DecisionTreeClassifier,GBTClassifier,RandomForestClassifier,MultilayerPerceptronClassifier

assembler = VectorAssembler(inputCols=
['Alcohol_10buckets',
 'BMI_10buckets',
 'Population_10buckets',
 'Schooling_10buckets',
 'expenditure_10buckets',
 'GDP_10buckets',                                      
 'income_10buckets'],outputCol='features')

DT = DecisionTreeClassifier(featuresCol='features',labelCol='life_expectancy_2buckets', maxBins = 120, maxDepth=12, impurity='entropy',)
pipeline = Pipeline(stages=[assembler,DT])
train, test = df.randomSplit([0.8,.2])
fit_model = pipeline.fit(train)
results = fit_model.transform(test)
results.select('features','rawPrediction','probability','prediction').show()
```

```

from pyspark.ml.evaluation import BinaryClassificationEvaluator

my_eval = BinaryClassificationEvaluator(rawPredictionCol='prediction',
                                       labelCol='life_expectancy_2buckets')
acc = my_eval.evaluate(results)
print(acc)
y_pred=results.select("prediction").collect()
y_orig=results.select("life_expectancy_2buckets").collect()

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_orig, y_pred)
print("Confusion Matrix:")
print(cm)
fit_model.stages[-1].featureImportances
```
